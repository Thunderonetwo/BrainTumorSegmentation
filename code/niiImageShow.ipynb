```python
import nibabel as nib
```

```python
epi_img = nib.load('Brats17_2013_10_1_t1.nii.gz')
epi_img_data = epi_img.get_data()
epi_img_data.shape
```

```python
import matplotlib.pyplot as plt



slice_0 = epi_img_data[:, :, 58]
slice_1 = epi_img_data[:, :, 59]
slice_2 = epi_img_data[:, :, 60]

slices = [slice_0, slice_1, slice_2]

fs = 100

fig, axs = plt.subplots(1, len(slices), figsize=(fs,fs))
for i, slice in enumerate(slices):
    axs[i].imshow(slice.T, cmap="gray", origin="lower")
plt.show()
plt.close()
```
