```python
import nibabel as nib
import matplotlib.pyplot as plt
```

```python
scan = nib.load('./prediction/validation_case_56/data_t1.nii.gz').get_data()
```

```python
segmentation = nib.load('./prediction/validation_case_56/prediction.nii.gz').get_data()
```

```python
label = nib.load('./prediction/validation_case_56/truth.nii.gz').get_data()
```

```python
scan.shape, segmentation.shape, label.shape
```

```python
#get the 59th layer
scanLayer = scan[:,:,80]
segLayer = segmentation[:,:,80]
labelLayer = label[:,:,80]
```

```python
scans = [scanLayer,segLayer,labelLayer]
```

```python
ax1 = plt.subplot(1,3,1)
ax1.imshow(scanLayer.T, origin = 'lower')
ax1.set_title("Brain Scan")
plt.axis('off')

ax2 = plt.subplot(1,3,2)
ax2.imshow(segLayer.T, origin = 'lower')
ax2.set_title("Prediction")
plt.axis('off')

ax3 = plt.subplot(1,3,3)
ax3.imshow(labelLayer.T, origin = 'lower')
ax3.set_title("Truth")
plt.axis('off')
plt.show()

```
